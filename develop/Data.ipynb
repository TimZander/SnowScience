```python
df_snow = pd.read_csv('../data/CAICVailSummitDangerScale.csv')
columns=['date', 'precip_accum_24_hour']
df_precip = pd.DataFrame(data=np.zeros((0,len(columns))), columns=columns)
size=len(df_snow)
for index, row in df_snow.iterrows():
    #print (row[0])
    sys.stdout.write("Download progress: %f   \r" % (index/size) )
    sys.stdout.flush()
    date=datetime.datetime.strptime(row[0], '%m/%d/%Y').strftime('%Y%m%d0000')
    #date_string='{0}{1}{2}0000'.format(date.year, date.month, date.day)
    #print(date)
    precip = m.attime(state='CO', county='Summit', vars='precip_accum_24_hour', attime=date, within='120')
    #if len(precip['STATION'])>1:
        #print(len(precip['STATION']))
    p=precip['STATION'][0]['OBSERVATIONS']['precip_accum_24_hour_value_1']['value']
    #print(p)
    df_precip = df_precip.append({'date':row[0], 'precip_accum_24_hour':p}, ignore_index=True)
#    break

df_combined = df_snow.set_index('date').join(df_precip.set_index('date'))
```

```python
from sklearn.cross_validation import train_test_split
X, Y = df.iloc[:,3:].values, df.iloc[:, 0].values
X_train, X_test, Y_train, Y_test = \
    train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
import sklearn.preprocessing as pp
#help(pp)
stdsc = pp.StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
LogisticRegression(penalty='l1')
lr=LogisticRegression(penalty='l1', C=6.8)
lr.fit(X_train_std, Y_train)
print('train accuracy: ', lr.score(X_train_std, Y_train))
print('test accuracy: ', lr.score(X_test_std, Y_test))
```
