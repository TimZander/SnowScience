```python
#Goal is to limit to just one observation a day. Air temp high and low we can likely just take from a specific slot
#perhaps the first

import pandas as pd
df_snow = pd.read_csv('../data/CAICVailSummitDangerScale.csv')
df_snow = df_snow.set_index('date')
```

```python
def fill_more_data(df, var):
    columns=['date', var]
    df_new = pd.DataFrame(data=np.zeros((0,len(columns))), columns=columns)
    size=len(df_snow)
    count=0
    for index, row in df.iterrows():
        sys.stdout.write("Download progress: %f   \r" % (count/size) )
        count=count+1
        sys.stdout.flush()
        try:
            date=datetime.datetime.strptime(index, '%m/%d/%Y').strftime('%Y%m%d0000')
            precip = m.attime(state='CO', county='Summit', vars=var, attime=date, within='120')
            sens_var=list(precip['STATION'][0]['SENSOR_VARIABLES'][var])[0]
            p=precip['STATION'][0]['OBSERVATIONS'][sens_var]['value']
            df_new = df_new.append({'date':index, var:p}, ignore_index=True)
        except:
            pass
    sys.stdout.write("\n")
    return df_new
```
